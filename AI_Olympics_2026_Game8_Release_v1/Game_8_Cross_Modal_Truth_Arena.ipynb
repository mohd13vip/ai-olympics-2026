# **🏟️ Game 8 — The Cross-Modal Truth Arena**
## When Image and Text Disagree, Who Should You Trust?

Your team has reached the final arena of Phase 1.

Until now, you have studied the data, cleaned it, trained models, optimized them, and explained their behavior.

Now the challenge becomes harder.

Each sample contains an image and a text.  
Sometimes both support the same truth.  
Sometimes the content is fake.  
Sometimes the image looks real, the text sounds believable, but the relationship between them is wrong.

Your mission is to build a multimodal system that decides whether the content is Real or Fake.

Do not treat image and text as two independent signals only.

The strongest systems will understand:

- what the image suggests,
- what the text suggests,
- whether the image and text support each other,
- which modality should be trusted more,
- and when the case should be sent for review.

Final Question:

Can your AI detect the truth behind both the image and the words?



---



## **Requirements**

Use the provided Game 8 files:

- game8_train.csv
- game8_validation.csv
- game8_public_test.csv

Your team must:

1. Build an image-based model or image representation.

2. Build a text-based model or text representation.

3. Build a fusion system that combines image and text.

4. Include a strategy for measuring or modeling the relationship between image and text.

5. Compare at least three systems:
   - Image-only
   - Text-only
   - Fusion-based

6. Explain how your system decides which signal is more important.

7. Generate predictions for the public test file.

8. Provide confidence scores for predictions.

9. Provide an evidence field explaining the main reason for the decision:
   - image
   - text
   - image_text_relation
   - uncertain

10. Optionally flag uncertain cases for human review.

11. Analyze errors on the validation set.

12. Save all required outputs in the correct format.

Important Rule:

The goal is not only high accuracy.

The goal is to build a system that can reason across modalities and justify its decision.
