# **🧪 Game 3 — The Noise Lab**

## Clean Carefully. Preserve What Matters.

Your team has completed the reconnaissance mission.

You now know that real-world data is not always consistent, complete, or equally reliable. However, preprocessing is not a checklist, and applying more transformations does not automatically produce a better model.

Some changes may improve the data.
Some may remove useful information.
Some samples may need repair.
Some may need special handling.
Others may need to remain untouched.

Your mission is to design a thoughtful preprocessing pipeline for both images and texts.

Do not apply transformations blindly.

Every preprocessing decision must be supported by evidence from the previous investigation and evaluated through controlled experiments.

### Final Question

> Can you improve data quality without destroying the signals your model needs?

## **🎯 Game 3 Requirements**

Use the dataset provided in the previous game.

Your team must:

1. Review the risks discovered during the reconnaissance mission.

2. Design a preprocessing pipeline for image data.

3. Design a preprocessing pipeline for text data.

4. Handle incomplete or unreliable samples using a clear strategy.

5. Explain which transformations were applied and why.

6. Explain which transformations were rejected and why.

7. Preserve the original data and create a processed version separately.

8. Compare multiple preprocessing choices through controlled experiments.

9. Use the same simple baseline before and after preprocessing.

10. Measure whether each preprocessing decision improves, harms, or does not affect the result.

11. Produce a final recommended pipeline for the next games.

### **Important Rule**

Do not use a larger model to hide weaknesses in the preprocessing strategy.

The goal is not to achieve the highest score yet.

The goal is to build a reliable data pipeline.
