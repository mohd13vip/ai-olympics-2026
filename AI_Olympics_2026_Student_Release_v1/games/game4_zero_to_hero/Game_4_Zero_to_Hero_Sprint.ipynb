# **🏃 Game 4 — The Zero-to-Hero Sprint**
Build Intelligence from the Ground Up.
Scenario

Your data pipeline is ready.

Now your team must prove that it understands how learning begins before using powerful pretrained systems.

You are not allowed to borrow intelligence yet.

Your mission is to build two models from scratch:

One model that understands images.
One model that understands texts.

The strongest solution is not necessarily the most complex one.

A simple model that is well designed, well trained, and honestly evaluated may reveal more than a large model built without understanding.

Final Question

How much can your system learn when it starts with no pretrained knowledge?

# **Requirements**

Use the processed Train and Validation files created in Game 3.

Your team must:

Build an image-classification model from scratch.
Build a text-classification model from scratch.
Train both models using the same Real/Fake objective.
Explain the architecture selected for each modality.
Record training and validation performance across epochs.
Visualize learning curves.
Evaluate both models using appropriate classification metrics.
Generate a confusion matrix for each model.
Compare image-only and text-only performance.
Analyze important error cases.
Explain the main limitations of both scratch models.
Save both trained models for comparison in the next game.
Important Rule

Do not use pretrained weights, pretrained encoders, or external trained representations.

The purpose of this game is to establish honest scratch baselines.
