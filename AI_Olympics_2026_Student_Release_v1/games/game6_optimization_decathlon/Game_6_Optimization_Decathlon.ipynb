# **⚙️ Game 6 — The Optimization Decathlon**
Improve with Evidence. Not Guesswork.
Scenario

Your models are working, but they are not finished.

You now enter the decathlon: a stage where many improvements are possible, but time and compute are limited.

Trying everything is not a strategy.

Your mission is to optimize your models through controlled experiments, where every change begins with a hypothesis and ends with a measured decision.

Some experiments may help.
Some may fail.
Some may improve one metric while harming another.

Your task is not to hide failed experiments.

Your task is to learn from them.

**Final Question**

Can your team improve performance through disciplined experimentation instead of random trial and error?



---



# **Requirements**

Use the strongest scratch and transfer models from the previous games.

Your team must:

Define a clear baseline for each modality.
Propose optimization hypotheses before running experiments.
Run controlled experiments where one major factor changes at a time.
Track the effect of each experiment.
Compare performance, training cost, and inference cost.
Record successful and unsuccessful experiments.
Explain why each experiment was accepted or rejected.
Select the final optimized image model.
Select the final optimized text model.
Export the final models for the explainability and fusion stages.
Produce an optimization history table.
Explain which optimization choices created the greatest value.

**Important Rule**

Do not report only your best result.

Your experiment history is part of the evaluation.

A failed experiment is acceptable when the reasoning and conclusion are clear.
